```python
# Step 1: Read the color image
img = cv2.imread("coin.jpg")
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # Convert BGR â†’ RGB for correct display
```

```python

# Step 2: Split into RGB components
R, G, B = cv2.split(img_rgb)
```

```python
# Step 3: Convert to Grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python

# Step 4: Convert to Binary (Thresholding)
# Pixels above 127 become 255 (white), others 0 (black)
_, binary = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)
```

```python
# Step 5: Display all results
plt.figure(figsize=(12, 8))

plt.subplot(2, 3, 1)
plt.imshow(img_rgb)
plt.title("Original Image")
plt.axis("off")

plt.subplot(2, 3, 2)
plt.imshow(R, cmap='Reds')
plt.title("Red Component")
plt.axis("off")

plt.subplot(2, 3, 3)
plt.imshow(G, cmap='Greens')
plt.title("Green Component")
plt.axis("off")

plt.subplot(2, 3, 4)
plt.imshow(B, cmap='Blues')
plt.title("Blue Component")
plt.axis("off")

plt.subplot(2, 3, 5)
plt.imshow(gray, cmap='gray')
plt.title("Grayscale Image")
plt.axis("off")

plt.subplot(2, 3, 6)
plt.imshow(binary, cmap='gray')
plt.title("Binary Image")
plt.axis("off")

plt.tight_layout()
plt.show()
```
